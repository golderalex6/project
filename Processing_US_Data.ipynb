```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import numpy as np
from math import *
mil_to_km=2.59
```

```python
us_raw=pd.read_csv('/content/us_data/covid19-us(main).csv')
us_raw.rename(columns={'State':'Abbreviation'},inplace=True)
us_raw.drop(index=0,inplace=True)
us_raw.reset_index(drop=True,inplace=True)
us_raw
```

```python
state_name=pd.read_csv('/content/us_data/state-acronym.csv')
state_name
```

```python
#HANDLE AREA
state_area=pd.read_csv('/content/us_data/us-state-areas.csv')
state_area['area (sq. km)']=state_area['area (sq. mi)']*mil_to_km
state_area
```

```python
#HANDLE POPULATION
state_population=pd.read_excel('/content/us_data/us-state-population.xlsx')
state_population=state_population[['Geographic Area',2021]]
state_population.drop(0,inplace=True)
state_population.reset_index(drop=True,inplace=True)
state_population['Geographic Area']=state_population['Geographic Area'].map(lambda x:x.split(', ')[1])
state_population=DataFrame.groupby(state_population,'Geographic Area').sum()
state_population.reset_index(inplace=True)
state_population.rename(columns={'Geographic Area':'State'},inplace=True)
state_population
```

# Density

```python
#HANDLE DENSITY
us_density=pd.merge(state_population,state_area,on='State')
us_density['density(peo/sq.km)']=us_density[2021]/us_density['area (sq. km)']
us_density.rename(columns={2021:'population'},inplace=True)
us_density
```

```python
filter_us_0=us_raw[[us_raw.columns[2]]+list(filter(lambda x:'2021-07-01'<=x<='2021-08-31',us_raw.columns))]
filter_us_1=us_raw[[us_raw.columns[2]]+list(filter(lambda x:'2021-07-02'<=x<='2021-09-01',us_raw.columns))]
```

# New case a Day

```python
#CREATE TABLE FOR GENERAL INFORMATION
a=DataFrame.groupby(filter_us_0,'Abbreviation').sum()
b=DataFrame.groupby(filter_us_1,'Abbreviation').sum()
us_case_day=DataFrame({a.columns[i]:b[b.columns[i]]-a[a.columns[i]] for i in range(len(a.columns))})
us_case_day[us_case_day<0]=0

us_case_2m=DataFrame(us_case_day.transpose().sum())
us_case_2m.rename(columns={0:'ncase_th_periods'},inplace=True)
us_case_2m.reset_index(inplace=True)
us_case_2m=pd.merge(us_case_2m,state_name,how='inner')[['State','ncase_th_periods']]
us_case_2m=pd.merge(us_case_2m,us_density,on='State')


us_case_2m_people=us_case_2m.sort_values('population')
us_case_2m_density=us_case_2m.sort_values('density(peo/sq.km)')
us_case_2m=DataFrame(us_case_2m[us_case_2m['State'].isin(['Pennsylvania','Arkansas','Maine','Virginia','New Hampshire','Georgia','Arizona','Texas','New York','North Dakota'])])
us_case_2m['AVG_case_per1000']=us_case_2m['ncase_th_periods']*1000/us_case_2m['population']
us_case_2m_people
```

# General Information

```python
#CREATE TABLE FOR TOTAL NEW CASE OF 10 STATE BY DAYS
u=DataFrame.groupby(filter_us_0,'Abbreviation').sum()
v=DataFrame.groupby(filter_us_1,'Abbreviation').sum()
us_case=DataFrame({u.columns[i]:v[v.columns[i]]-u[u.columns[i]] for i in range(len(u.columns))})
us_case[us_case<0]=0

us_day=us_case.reset_index()
us_day=pd.merge(us_day,state_name,how='inner')
us_day.drop('Abbreviation',axis=1,inplace=True)
us_day_2m=us_day[us_day['State'].isin(['Pennsylvania','Arkansas','Maine','Virginia','New Hampshire','Georgia','Arizona','Texas','New York','North Dakota'])]
us_day_2m.set_index('State',inplace=True)
us_day_2m
```
